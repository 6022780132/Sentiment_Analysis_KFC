# data Clearing and Collecting

```python
# Clearing the columns name
df.columns = df.columns.str.replace('ความชอบในแต่ละเมนูของ KFC ',r'')
df.columns = df.columns.str.replace('Fried Chicken - ',r'')
df.columns = df.columns.str.replace('[',r'')
df.columns = df.columns.str.replace(']',r'')
df.columns = df.columns.str.replace('(','-')
df.columns = df.columns.str.replace(')',r'')
#df.columns = df.columns[:11].union(df.columns[11:].str.lower())
df.columns
```

```python
#collect score for each menu
#count[menu(0,28)][satisfaction level (0,3)]
#count_score[menu(0,28)]

count = np.zeros((29,4))
count_score = [None]*29

for i in range(len(df.columns)):
    
    count_vector=df[df.columns[i]].value_counts()
    
    if 'ไม่ชอบ' in count_vector : count[i][0] = count_vector.ไม่ชอบ
    if 'เฉยๆ' in count_vector : count[i][1] = count_vector.เฉยๆ
    if 'ชอบ' in count_vector : count[i][2]= count_vector.ชอบ
    if 'ชอบมาก' in count_vector : count[i][3] = count_vector.ชอบมาก
    print('for i = ' + str(i) + ' : ' + str(df.columns[i]) )
    print(count[i][0],count[i][1],count[i][2],count[i][3])

    count_score[i] = [sum([(count[i][x])*(x+1) for x in range(0,4)])]
    print('count_score = ' + str(count_score[i]))
    print()
```

```python
t = [None]*6

#calulate t value
#input w h a i
while True :
    default = input('defalut value (y/n) = ')
    if default.lower() =='y' :
        w = 60
        h = 170
        a = 21
        i = 2
        sex = 'male'
        break
    elif default.lower() =='n' :
        w = input('weight(kg) = ')
        h = input('height(cm) = ')
        a = input('age = ')
        i = input('active level [1:4] = ') #i=1,2,3,4
        sex = input('sex {male,female} = ')
        break
    elif default.lower() =='view' :
        w = 80
        h = 180
        a = 21
        i = 4
        sex = 'male'
        break
    else :
        print('try to input correct alphabet : ')

level = [1.20,1.45,1.70,1.95]
bmr = 10*int(w) + 6.25*int(h) - 5*int(a) + 5
if sex.lower()=='male':
    nsex = 200
elif sex.lower()=='female':
    nsex = 0
t[0] = nsex+bmr*level[int(i)-1] #calories
#t[1] = t[0]*0.3/9 #fat(g)
#t[2] = t[0]*(0.4+(int(i)*0.05))/4 #carb(g)
#t[3] = (t[0]-((t[1]*9)+(t[2]*4)))/4 #protein(g)
t[1] = t[0]*0.4/9 #fat(g)
t[2] = t[0]*0.4/4 #carb(g)
t[3] = t[0]*0.2/4 #protein(g)
t[4] = 2000 #sodium
t[5] = 1000 #cholesterol
print('total calories = ' + str(t[0]))
print('protein = ' + str(t[3]) + ' , fat = ' + str(t[1]) + ' , carb = ' + str(t[2]))
print('sodium = ' + str(t[4]) + ' , cholesterol = ' + str(t[5]))
```

```python
#df target
dftarget = pd.DataFrame(t).T
dftarget = dftarget.rename(index={dftarget.index[0]:'target'})
for x in range(len(dftarget.columns)):
    dftarget = dftarget.rename(columns={dftarget.columns[x]:df3.columns[x+1]})
dftarget
```

# selection

```python
limit = 10
limitin = 20
i = 0
k = 0
sample = 200
temp = pd.DataFrame()
stack = pd.DataFrame()
stest = pd.DataFrame()
sdfcon = pd.DataFrame()
sdfdiff = pd.DataFrame()
result = pd.DataFrame()
serror = pd.DataFrame()
```

```python
def start() :
    global temp
    global stack
    global stest
    global sdfcon
    global sdfdiff
    global result
    global i
    temp = 0
    stack = pd.DataFrame()
    stest = pd.DataFrame()
    sdfcon = pd.DataFrame()
    sdfdiff = pd.DataFrame()
    result = pd.DataFrame()
    i = 0
    newmenu()
    

#menu
def newmenu() :
    
    global dfnewmenu
    global prop
    global stest
    global stack
    global sdfcon
    
    if i > sample-sample*0.2 :
        stest = df3.filter(items=[random.randint(29, 31)],axis=0)
        print('i>sample-10')
    else :
        findnewmenu()
        stest = dfnewmenu #need better random with weighting factor
    
    stesti = pd.DataFrame(stest.iloc[0,1:7]).T
    stack = pd.concat([stack,stesti])
    stack = stack.rename(index={stack.index[-1]:stest.iloc[0,0]})
    print()
    print('new stack having : '+str(len(stack.index)))
    #sdfcon or total consumtion
    stotalcon = 0
    for x in range(len(stack.index)) :
        stotalcon += (stack.iloc[x,:])
    sdfcon = pd.DataFrame(stotalcon).T
    sdfcon = sdfcon.rename(index={sdfcon.index[0]:'total consume'})
    diff()
    
    
def eliminate() :
    global stack
    global i
    stack = stack.head(len(stack.index)-1)
    i += 1
    print('stack -1')
    print('i = '+str(i))
    


def diff() : #just find diff not checking
    global sdfcon
    global dftarget
    global sdfdiff
    global serror
    global errormean
    global i
    global sample
    #sdiff
    sdiff = dftarget.iloc[0,:]-sdfcon.iloc[0,:]
    sdfdiff = pd.DataFrame(sdiff).T
    sdfdiff = sdfdiff.rename(index={sdfdiff.index[0]:'different'})
    #serror
    serror = sdfdiff*100/(dftarget.iloc[0,:])
    serror = serror.rename(index={serror.index[0]:'error'})
    #mean
    errormean = sum(abs(serror.iloc[0,serror.columns != 'Sodium (mg)']))/((len(serror.columns[:]))-1) #might be[:-2]
    #get new error mean but still not confirm.
    
    if temp == 0 :
        print('temp = 0')
        checkerror()
    else :
        print('having temp')
        if i<sample :
            print('go to checkmean2temp')
            checkmean2temp()
            print('done check2temp')
        else :
            print('enough , i is maximum.')
            output()
            print()
            print('w a r n i n g - w a r n i n g - w a r n i n g')
            print()
        

# id error > limit
def checkerror():
    global serror
    global limitin
    global temp
    global errormean
    print('new menu = ' + str(stest.iloc[0,0]))
    for x in range(len(serror.columns)) :
        print('x = ' + str(x))
        if serror.iloc[0,x]>limitin : #macro is lack then newmenu
            print(str(serror.columns[x])+' is outside limitin : error(%) = '+str(serror.iloc[0,x]))
            temp = errormean
            print('temp saved')
            print('break to new menu')
            newmenu()
            print('break from check error')
            break
        elif abs(serror.iloc[0,x])>20 : #macro is surplus then stack-1 , newmenu
            print(str(serror.columns[x])+' is outside limitin : error(%) = '+str(serror.iloc[0,x]))
            print('go to eliminate')
            eliminate()
            print('eliminated then... newmenu')
            newmenu()
            print('break from checkerror')
            break
        else :
            print(str(serror.columns[x])+' is inside limitin : error(%) = '+str(serror.iloc[0,x]))
            if x == 5 :
                checkmean()
                print('nice job , every macro is in limitin. Then checkmean')

                
def checkmean() :
    global errormean
    global limit
    global temp
    print('error mean = '+ str(errormean))
    if errormean < limit :
        print( 'error mean is less than limit , lets do output' )
        output()
    else :
        print('error mean is too large , fine , save error in temp')
        temp = errormean
        print('temp saved')
        newmenu()          
        
#check if this new stack has less diff
def checkmean2temp () :
    global temp
    global i
    global errormean
    global k
    if abs(errormean)<abs(temp) :
        print('new errormean < temp(oldmean) : '+str(errormean)) 
        #confirm errormean is better
        print('go to check error')
        checkerror()
        print('done checkerror')
    else :
        print('This errormean > temp(olderror) , eliminate it and newmenu')
        k += 1
        if k == 3 :
            print('double eliminate')
            eliminate()
        eliminate()
        print('done eliminate')
        print('go newmenu')
        newmenu()
        print('done newmenu')

def output():
    global stack
    global sdfcon
    global serror
    global sdfdiff
    global dftarget
    global result
    global errormean
    result = pd.concat([dftarget,sdfcon,sdfdiff,serror,stack])
    print('Average error is '+str(errormean))
```
